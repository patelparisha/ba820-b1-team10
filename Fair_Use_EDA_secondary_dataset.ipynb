## 1) Variation in Fair Use Outcomes and Judicial Reasoning Across Courts

```python
# Fair use success rate by jurisdiction
cases_df.groupby("jurisdiction")["fair_use_found"].mean().sort_values(ascending=False)

# Case volume by jurisdiction (to assess coverage)
cases_df["jurisdiction"].value_counts().head(10)

# Average holding length by court (proxy for reasoning depth/style)
findings_df.groupby("court")["holding"].apply(
    lambda x: x.str.split().map(len).mean()
).sort_values(ascending=False).head(10)

```

## 2) Recurring Legal Language in Fair Use Reasoning

```python
# How often does transformative language appear?
findings_df["holding"].str.contains("transform", case=False).mean()

# Related recurring legal language
findings_df["holding"].str.contains("purpose and character", case=False).mean()
findings_df["holding"].str.contains("market", case=False).mean()

# Extract simple phrase windows around 'transform'
def extract_transform_context(text, window=8):
    tokens = text.lower().split()
    idxs = [i for i,t in enumerate(tokens) if "transform" in t]
    contexts = []
    for i in idxs:
        contexts.append(" ".join(tokens[max(0,i-window):min(len(tokens), i+window)]))
    return contexts

transform_phrases = findings_df["holding"].apply(extract_transform_context).explode()
transform_phrases.value_counts().head(10)
```

## 3) Dominant Content Categories in Fair Use Cases

```python
import pandas as pd
import matplotlib.pyplot as plt

CAT_COL = "categories"
TOP_N = 12

cats = (
    cases_df[CAT_COL]
    .dropna()
    .astype(str)
    .str.split(";")
    .explode()
    .str.strip()
)
cats = cats[cats != ""]

cat_counts = cats.value_counts()
top_counts = cat_counts.head(TOP_N)

plt.figure(figsize=(10, 6))
top_counts.sort_values(ascending=True).plot(kind="barh")
plt.xlabel("Number of cases (category appearances)")
plt.ylabel("Category")
plt.title(f"Top {TOP_N} content categories in fair use cases")
plt.tight_layout()
plt.show()

display(top_counts.rename_axis("category").reset_index(name="count"))
```

## 4) Fair Use Outcomes Across Jurisdictions

```python
jur_rate = (
    cases_df.groupby("jurisdiction")["fair_use_found"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top10 = jur_rate.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10.index.astype(str), top10.values * 100)  # convert to %
plt.ylabel("Fair use success rate (%)")
plt.xlabel("Jurisdiction")
plt.title("Top 10 jurisdictions by fair use success rate")
plt.xticks(rotation=45, ha="right")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

## 5) Number of Content Categories per Fair Use Case

```python
# Count number of categories per case
cases_df["num_categories"] = cases_df["categories"].str.split(";").apply(len)

# Count how many cases fall into each # of categories
dist = cases_df["num_categories"].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(dist.index.astype(int), dist.values)
plt.xlabel("Number of categories per case")
plt.ylabel("Number of cases")
plt.title("Distribution of categories per case")
plt.xticks(dist.index.astype(int))
plt.tight_layout()
plt.show()
```

## 6) Average Length of Judicial Holdings by Court

```python
avg_len = (
    findings_df.groupby("court")["holding"]
    .apply(lambda x: x.str.split().map(len).mean())
    .sort_values(ascending=False)
    .head(10)
)

avg_len.plot(kind="bar")

plt.title("Average Holding Length by Court (Top 10)")
plt.xlabel("Court")
plt.ylabel("Average Holding Length (words)")
plt.xticks(rotation=75, ha="right")
plt.tight_layout()
plt.show()
```

## 7)Length and Variability of Issue vs. Holding Text

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute word counts (robust to NaNs)
findings_df["issue_wc"] = (
    findings_df["issue"]
    .fillna("")
    .astype(str)
    .str.split()
    .str.len()
)

findings_df["holding_wc"] = (
    findings_df["holding"]
    .fillna("")
    .astype(str)
    .str.split()
    .str.len()
)

# drop rows where both are 0
plot_df = findings_df[["issue_wc", "holding_wc"]].copy()
plot_df = plot_df[~((plot_df["issue_wc"] == 0) & (plot_df["holding_wc"] == 0))]

# Side-by-side boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(
    [plot_df["issue_wc"].values, plot_df["holding_wc"].values],
    labels=["Issue text", "Holding text"],
    showfliers=False
)
plt.ylabel("Word count")
plt.title("Issue vs. Holding word count distributions")
plt.tight_layout()
plt.show()
```

## 8) Common Legal Concepts Referenced in Fair Use Holdings

```python
terms = {
    "Transformative": findings_df["holding"].str.contains("transform", case=False).mean(),
    "Purpose & Character": findings_df["holding"].str.contains("purpose and character", case=False).mean(),
    "Market Harm": findings_df["holding"].str.contains("market", case=False).mean()
}

plt.bar(terms.keys(), terms.values())

plt.title("Prevalence of Key Fair Use Reasoning Concepts")
plt.ylabel("Proportion of Cases")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
