```python
# ============================================================
# Hollywood Age Gap — EDA (Aligned to Your 4 Domain Questions)
# Data files (uploaded):
# "/content/age_gaps.csv"
# "/content/post_vars.json"
# "/content/readme-5.md"
# ============================================================

import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Optional network tools (fallback code included below if not installed)
try:
    import networkx as nx
    HAS_NETWORKX = True
except ImportError:
    HAS_NETWORKX = False

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)

DATA_PATH = "/content/age_gaps.csv"
POST_VARS_PATH = "/content/post_vars.json"
README_PATH = "/content/readme-5.md"

# -----------------------------
# 0) Load ancillary project info (source link, alt-text)
# -----------------------------
post_vars = {}
if os.path.exists(POST_VARS_PATH):
    with open(POST_VARS_PATH, "r", encoding="utf-8") as f:
        post_vars = json.load(f)

print("Project source link (if provided):", post_vars.get("article_link"))
print("Alt text keys found:", [k for k in post_vars.keys() if k.endswith("_alt")])

# -----------------------------
# 1) Load data + quick structure checks (required EDA baseline)
# -----------------------------
df = pd.read_csv(DATA_PATH)

print("\n=== Basic Shape ===")
print("Rows, Cols:", df.shape)
print("\n=== Columns ===")
print(list(df.columns))

print("\n=== dtypes ===")
print(df.dtypes)

print("\n=== Missing values ===")
print(df.isna().sum().sort_values(ascending=False).head(20))

print("\n=== Duplicate rows (entire row) ===")
print(df.duplicated().sum())

# -----------------------------
# 2) Minimal cleaning / type fixes + derived fields
# -----------------------------
# Dates
df["actor_1_birthdate"] = pd.to_datetime(df["actor_1_birthdate"], errors="coerce")
df["actor_2_birthdate"] = pd.to_datetime(df["actor_2_birthdate"], errors="coerce")

# Consistent string trimming
for c in ["movie_name", "director", "actor_1_name", "actor_2_name", "character_1_gender", "character_2_gender"]:
    df[c] = df[c].astype(str).str.strip()

# Sanity check: age_difference should equal actor_1_age - actor_2_age (older - younger)
df["calc_age_diff"] = df["actor_1_age"] - df["actor_2_age"]
mismatch = (df["calc_age_diff"] != df["age_difference"]).sum()
print("\n=== Sanity check: computed age diff mismatches ===")
print("Mismatched rows:", mismatch)

# Decade field for Q1
df["decade"] = (df["release_year"] // 10) * 10

# Pairing type (based on character gender labels in dataset)
df["pairing"] = df["character_1_gender"].str.lower() + "-" + df["character_2_gender"].str.lower()

# Large-gap flag used in multiple analyses (tune threshold if you want)
LARGE_GAP = 20
df["is_large_gap"] = df["age_difference"] >= LARGE_GAP

print("\n=== Range checks ===")
print("Release year:", df["release_year"].min(), "→", df["release_year"].max())
print("Age difference summary:\n", df["age_difference"].describe())

# -----------------------------
# 3) Dataset description helpers (for your writeup section)
# -----------------------------
def dataset_profile_table(df: pd.DataFrame) -> pd.DataFrame:
    out = []
    for c in df.columns:
        out.append({
            "column": c,
            "dtype": str(df[c].dtype),
            "n_unique": df[c].nunique(dropna=True),
            "missing": int(df[c].isna().sum()),
            "example_values": ", ".join(map(str, df[c].dropna().unique()[:3]))
        })
    return pd.DataFrame(out).sort_values(["missing", "n_unique"], ascending=[False, False])

profile = dataset_profile_table(df)
print("\n=== Column profile (top 15 by uniqueness) ===")
print(profile.sort_values("n_unique", ascending=False).head(15))

print("\n=== Stakeholder / domain-friendly quick stats ===")
print("Unique movies (by name):", df["movie_name"].nunique())
print("Unique directors:", df["director"].nunique())
print("Unique older-actor names:", df["actor_1_name"].nunique())
print("Unique younger-actor names:", df["actor_2_name"].nunique())
print("Couples per (movie_name, release_year) summary:")
print(df.groupby(["movie_name", "release_year"]).size().describe())

# ============================================================
# Q1) How has the “typical” age gap shifted by decade, and inflection points?
# ============================================================

decade_stats = (
    df.groupby("decade")
      .agg(
          n_pairs=("age_difference", "size"),
          mean_gap=("age_difference", "mean"),
          median_gap=("age_difference", "median"),
          p_large_gap=("is_large_gap", "mean"),
          p_mw=("pairing", lambda s: np.mean(s == "man-woman")),
          p_wm=("pairing", lambda s: np.mean(s == "woman-man")),
          p_mm=("pairing", lambda s: np.mean(s == "man-man")),
          p_ww=("pairing", lambda s: np.mean(s == "woman-woman")),
      )
      .reset_index()
      .sort_values("decade")
)

# Inflection points: biggest decade-to-decade changes in median
decade_stats["delta_median"] = decade_stats["median_gap"].diff()
decade_stats["abs_delta_median"] = decade_stats["delta_median"].abs()

print("\n=== Q1: Decade summary ===")
print(decade_stats)

top_inflections = decade_stats.dropna().sort_values("abs_delta_median", ascending=False).head(5)
print("\n=== Q1: Largest inflection points (by |Δ median|) ===")
print(top_inflections[["decade", "median_gap", "delta_median", "n_pairs"]])

# Plot: median and mean over decades + % large gaps
fig = plt.figure(figsize=(11, 5))
plt.plot(decade_stats["decade"], decade_stats["median_gap"], marker="o", label="Median age gap")
plt.plot(decade_stats["decade"], decade_stats["mean_gap"], marker="o", label="Mean age gap")
plt.title("Q1: Typical on-screen age gap by decade")
plt.xlabel("Decade")
plt.ylabel("Age difference (years)")
plt.grid(True, alpha=0.25)
plt.legend()
plt.show()

fig = plt.figure(figsize=(11, 4))
plt.plot(decade_stats["decade"], decade_stats["p_large_gap"] * 100, marker="o")
plt.title(f"Q1: Share of couples with large age gaps (≥ {LARGE_GAP} years), by decade")
plt.xlabel("Decade")
plt.ylabel("% large-gap pairings")
plt.grid(True, alpha=0.25)
plt.show()

# Optional: distribution snapshots (boxplots by decade groupings)
# (Use bins to avoid too many decades with low counts)
df["era"] = pd.cut(
    df["release_year"],
    bins=[1929, 1959, 1979, 1999, 2009, 2022],
    labels=["1930s-1950s", "1960s-1970s", "1980s-1990s", "2000s", "2010s-2022"],
    right=True
)

fig = plt.figure(figsize=(11, 5))
era_order = ["1930s-1950s", "1960s-1970s", "1980s-1990s", "2000s", "2010s-2022"]
data = [df.loc[df["era"] == e, "age_difference"].values for e in era_order]
plt.boxplot(data, labels=era_order, showfliers=True)
plt.title("Q1: Age-gap distribution by era (boxplot)")
plt.xlabel("Era")
plt.ylabel("Age difference (years)")
plt.grid(True, axis="y", alpha=0.25)
plt.show()

# ============================================================
# Q2) Do directors cluster into “casting age-gap profiles”, and stability over career?
# ============================================================

# Director-level features (keep only directors with enough observations for stable profiling)
MIN_COUPLES_PER_DIRECTOR = 5

dir_base = (
    df.groupby("director")
      .agg(
          n_pairs=("age_difference", "size"),
          mean_gap=("age_difference", "mean"),
          median_gap=("age_difference", "median"),
          std_gap=("age_difference", "std"),
          p_large_gap=("is_large_gap", "mean"),
          p_mw=("pairing", lambda s: np.mean(s == "man-woman")),
          p_wm=("pairing", lambda s: np.mean(s == "woman-man")),
          p_mm=("pairing", lambda s: np.mean(s == "man-man")),
          p_ww=("pairing", lambda s: np.mean(s == "woman-woman")),
          first_year=("release_year", "min"),
          last_year=("release_year", "max"),
      )
      .reset_index()
)

dir_feat = dir_base[dir_base["n_pairs"] >= MIN_COUPLES_PER_DIRECTOR].copy()
dir_feat["career_span_years"] = dir_feat["last_year"] - dir_feat["first_year"]

print("\n=== Q2: Directors retained for profiling ===")
print("Directors total:", dir_base.shape[0])
print(f"Directors with >= {MIN_COUPLES_PER_DIRECTOR} couples:", dir_feat.shape[0])

feature_cols = ["mean_gap", "median_gap", "std_gap", "p_large_gap", "p_mw", "p_wm", "p_mm", "p_ww"]
X = dir_feat[feature_cols].fillna(0.0).values

scaler = StandardScaler()
Xz = scaler.fit_transform(X)

# Pick k via silhouette (small search)
k_candidates = list(range(2, 7))  # tune if desired
sil_scores = {}
for k in k_candidates:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = km.fit_predict(Xz)
    sil_scores[k] = silhouette_score(Xz, labels)

best_k = max(sil_scores, key=sil_scores.get)
print("\n=== Q2: Silhouette scores (director clustering) ===")
print(sil_scores)
print("Chosen k:", best_k)

km = KMeans(n_clusters=best_k, random_state=42, n_init="auto")
dir_feat["cluster"] = km.fit_predict(Xz)

# Cluster interpretation table
cluster_summary = (
    dir_feat.groupby("cluster")
            .agg(
                n_directors=("director", "size"),
                avg_pairs=("n_pairs", "mean"),
                mean_gap=("mean_gap", "mean"),
                median_gap=("median_gap", "mean"),
                p_large_gap=("p_large_gap", "mean"),
                p_mw=("p_mw", "mean"),
                p_wm=("p_wm", "mean"),
                p_mm=("p_mm", "mean"),
                p_ww=("p_ww", "mean"),
                avg_span=("career_span_years", "mean"),
            )
            .reset_index()
            .sort_values("mean_gap", ascending=False)
)

print("\n=== Q2: Cluster summaries (interpret as 'profiles') ===")
print(cluster_summary)

# Visual: mean_gap vs p_large_gap for clusters
fig = plt.figure(figsize=(9, 5))
for cl in sorted(dir_feat["cluster"].unique()):
    sub = dir_feat[dir_feat["cluster"] == cl]
    plt.scatter(sub["mean_gap"], sub["p_large_gap"] * 100, label=f"Cluster {cl}", alpha=0.7)
plt.title("Q2: Director profiles — Mean age gap vs % large-gap pairings")
plt.xlabel("Director mean age gap")
plt.ylabel(f"% pairings with gap ≥ {LARGE_GAP}")
plt.grid(True, alpha=0.25)
plt.legend()
plt.show()

# Stability over career:
# For each director, compute early vs late career gap stats (split by midpoint year)
def early_late_stability(df_dir: pd.DataFrame) -> pd.Series:
    years = df_dir["release_year"].values
    mid = np.median(years)
    early = df_dir[df_dir["release_year"] <= mid]
    late = df_dir[df_dir["release_year"] > mid]
    if len(early) < 2 or len(late) < 2:
        return pd.Series({"delta_mean": np.nan, "delta_median": np.nan, "delta_p_large": np.nan})
    return pd.Series({
        "delta_mean": late["age_difference"].mean() - early["age_difference"].mean(),
        "delta_median": late["age_difference"].median() - early["age_difference"].median(),
        "delta_p_large": late["is_large_gap"].mean() - early["is_large_gap"].mean(),
    })

stability = (
    df.merge(dir_feat[["director", "cluster"]], on="director", how="inner")
      .groupby(["director", "cluster"], as_index=False)
      .apply(lambda g: early_late_stability(g))
      .reset_index(drop=True)
)

print("\n=== Q2: Early vs late career stability (sample) ===")
print(stability.head(10))

# Plot distribution of stability deltas by cluster
fig = plt.figure(figsize=(10, 4))
for cl in sorted(stability["cluster"].unique()):
    sub = stability[stability["cluster"] == cl]["delta_median"].dropna()
    plt.hist(sub, bins=20, alpha=0.5, label=f"Cluster {cl}")
plt.title("Q2: Change in director median gap (late - early), by cluster")
plt.xlabel("Δ median age gap (years)")
plt.ylabel("Count of directors")
plt.grid(True, alpha=0.2)
plt.legend()
plt.show()

# ============================================================
# Q3) Multiple-couple movies: headline pairing vs systematic within-film pattern
# ============================================================

movie_key = ["movie_name", "release_year"]

movie_stats = (
    df.groupby(movie_key)
      .agg(
          n_couples=("age_difference", "size"),
          max_gap=("age_difference", "max"),
          min_gap=("age_difference", "min"),
          mean_gap=("age_difference", "mean"),
          median_gap=("age_difference", "median"),
          std_gap=("age_difference", "std"),
      )
      .reset_index()
)

movie_stats["range_gap"] = movie_stats["max_gap"] - movie_stats["min_gap"]
movie_stats["headline_share"] = movie_stats["max_gap"] / movie_stats["age_difference_sum"] if "age_difference_sum" in movie_stats else np.nan

# Better "headline share": max gap / sum of gaps within movie
gap_sum = df.groupby(movie_key)["age_difference"].sum().reset_index(name="sum_gap")
movie_stats = movie_stats.merge(gap_sum, on=movie_key, how="left")
movie_stats["headline_share"] = movie_stats["max_gap"] / movie_stats["sum_gap"]

multi = movie_stats[movie_stats["n_couples"] > 1].copy()
single = movie_stats[movie_stats["n_couples"] == 1].copy()

print("\n=== Q3: Movies with multiple couples ===")
print("Total movie-year entries:", movie_stats.shape[0])
print("Multi-couple movie-year entries:", multi.shape[0])
print("Max couples in a movie-year:", movie_stats["n_couples"].max())

print("\n=== Q3: Within-movie variance metrics (multi-couple only) ===")
print(multi[["n_couples", "max_gap", "min_gap", "range_gap", "std_gap", "headline_share"]].describe())

# Plot: distribution of within-movie range (multi-couple)
fig = plt.figure(figsize=(10, 4))
plt.hist(multi["range_gap"].dropna(), bins=25)
plt.title("Q3: Within-movie age-gap range (max - min) for multi-couple movies")
plt.xlabel("Range of age gaps within same film (years)")
plt.ylabel("Number of films")
plt.grid(True, alpha=0.2)
plt.show()

# Plot: headline share (is there one dominant pairing?)
fig = plt.figure(figsize=(10, 4))
plt.hist(multi["headline_share"].dropna(), bins=25)
plt.title("Q3: Headline pairing dominance (max gap / sum of gaps) in multi-couple movies")
plt.xlabel("Headline share (closer to 1 = one pairing dominates)")
plt.ylabel("Number of films")
plt.grid(True, alpha=0.2)
plt.show()

# Identify extreme examples to discuss in writeup
print("\n=== Q3: Examples where one pairing dominates (headline_share >= 0.70) ===")
print(
    multi[multi["headline_share"] >= 0.70]
    .sort_values("headline_share", ascending=False)
    .head(15)[movie_key + ["n_couples", "max_gap", "min_gap", "range_gap", "headline_share"]]
)

print("\n=== Q3: Examples where gaps are more uniform (range_gap <= 3) ===")
print(
    multi[multi["range_gap"] <= 3]
    .sort_values(["n_couples", "range_gap"], ascending=[False, True])
    .head(15)[movie_key + ["n_couples", "max_gap", "min_gap", "range_gap", "headline_share"]]
)

# ============================================================
# Q4) Actors in large-gap pairings: repeated "hubs" + network structure
# ============================================================

# Actor-level recurrence (older vs younger roles)
older_actor_stats = (
    df.groupby("actor_1_name")
      .agg(
          n_pairs=("age_difference", "size"),
          mean_gap=("age_difference", "mean"),
          p_large=("is_large_gap", "mean"),
          median_older_age=("actor_1_age", "median"),
      )
      .reset_index()
      .rename(columns={"actor_1_name": "actor_name"})
)
older_actor_stats["role"] = "older_partner"

younger_actor_stats = (
    df.groupby("actor_2_name")
      .agg(
          n_pairs=("age_difference", "size"),
          mean_gap=("age_difference", "mean"),
          p_large=("is_large_gap", "mean"),
          median_younger_age=("actor_2_age", "median"),
      )
      .reset_index()
      .rename(columns={"actor_2_name": "actor_name"})
)
younger_actor_stats["role"] = "younger_partner"

actor_role_stats = pd.concat([older_actor_stats, younger_actor_stats], ignore_index=True)

# Top recurring large-gap actors in each role
print("\n=== Q4: Top 'older partner' large-gap hubs (min 5 pairings) ===")
print(
    older_actor_stats[older_actor_stats["n_pairs"] >= 5]
    .sort_values(["p_large", "n_pairs"], ascending=[False, False])
    .head(20)
)

print("\n=== Q4: Top 'younger partner' large-gap hubs (min 5 pairings) ===")
print(
    younger_actor_stats[younger_actor_stats["n_pairs"] >= 5]
    .sort_values(["p_large", "n_pairs"], ascending=[False, False])
    .head(20)
)

# Network: edges = (older_actor -> younger_actor) for large-gap pairings
large_edges = df[df["is_large_gap"]][["actor_1_name", "actor_2_name", "age_difference", "movie_name", "release_year"]].copy()

print("\n=== Q4: Large-gap edge list ===")
print("Edges:", large_edges.shape[0], "out of total pairings:", df.shape[0])

if HAS_NETWORKX:
    G = nx.DiGraph()
    for _, r in large_edges.iterrows():
        u = r["actor_1_name"]
        v = r["actor_2_name"]
        # weight counts co-appearances; keep max gap as attribute
        if G.has_edge(u, v):
            G[u][v]["weight"] += 1
            G[u][v]["max_gap"] = max(G[u][v]["max_gap"], int(r["age_difference"]))
        else:
            G.add_edge(u, v, weight=1, max_gap=int(r["age_difference"]))

    # Hubs:
    out_deg = pd.Series(dict(G.out_degree(weight="weight"))).sort_values(ascending=False)
    in_deg = pd.Series(dict(G.in_degree(weight="weight"))).sort_values(ascending=False)

    print("\n=== Q4: Network hubs (older-side out-degree, weighted) ===")
    print(out_deg.head(20))

    print("\n=== Q4: Network hubs (younger-side in-degree, weighted) ===")
    print(in_deg.head(20))

    # Optional: simple visualization of top subgraph (small to keep readable)
    TOP_N = 25
    top_nodes = set(out_deg.head(TOP_N).index) | set(in_deg.head(TOP_N).index)
    H = G.subgraph(top_nodes).copy()

    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(H, seed=42, k=0.6)
    nx.draw_networkx_nodes(H, pos, node_size=200, alpha=0.8)
    nx.draw_networkx_edges(H, pos, alpha=0.25, arrows=True)
    nx.draw_networkx_labels(H, pos, font_size=8)
    plt.title(f"Q4: Large-gap casting network (top hubs; ≥{LARGE_GAP} year gaps)")
    plt.axis("off")
    plt.show()

else:
    # Fallback (no networkx): compute weighted degrees from edge list
    # "older hubness": how often actor_1 appears in large-gap edges
    out_deg = large_edges.groupby("actor_1_name").size().sort_values(ascending=False)
    in_deg = large_edges.groupby("actor_2_name").size().sort_values(ascending=False)

    print("\n[networkx not installed] Fallback hub tables")
    print("\n=== Q4: Older-side hubs (count of large-gap pairings) ===")
    print(out_deg.head(20))
    print("\n=== Q4: Younger-side hubs (count of large-gap pairings) ===")
    print(in_deg.head(20))

# ============================================================
# Extra EDA: Correlation/associations that motivate your questions
# (Keep these outputs for your "motivated by EDA" writeup)
# ============================================================

# Relationship between younger actor age and gap size (is large gap mostly driven by very young partner?)
fig = plt.figure(figsize=(10, 4))
plt.scatter(df["actor_2_age"], df["age_difference"], alpha=0.35)
plt.title("EDA: Age difference vs younger actor age")
plt.xlabel("Younger actor age at release")
plt.ylabel("Age difference (years)")
plt.grid(True, alpha=0.2)
plt.show()

corr = df[["actor_1_age", "actor_2_age", "age_difference", "release_year"]].corr(numeric_only=True)
print("\n=== EDA: Correlation matrix ===")
print(corr)

# Pairing-type differences (character gender labels)
pair_stats = (
    df.groupby("pairing")
      .agg(
          n=("age_difference", "size"),
          mean_gap=("age_difference", "mean"),
          median_gap=("age_difference", "median"),
          p_large=("is_large_gap", "mean"),
          mean_older_age=("actor_1_age", "mean"),
          mean_younger_age=("actor_2_age", "mean"),
      )
      .reset_index()
      .sort_values("n", ascending=False)
)
print("\n=== EDA: Summary by pairing type ===")
print(pair_stats)

# ------------------------------------------------------------
# End of EDA script
# ------------------------------------------------------------
```
